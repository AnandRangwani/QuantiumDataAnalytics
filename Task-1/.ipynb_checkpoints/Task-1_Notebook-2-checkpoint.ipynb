```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Cleaned_merged_data.csv',index_col=0)
df.head()
```

### To find out which product is expensive and which one is relatively cheaper , we need to calculate price per 100 grams of product 

```python
df['PricePer100g']=((df['TOT_SALES']/df['PROD_QTY'])*100)/df['PROD_WEIGHT']
```

```python
df.head()
```

# We will try to find out that what category of customers are of greater concern for the supermarket , and will try to analyse their choice of flavours of chips.



### For a super-market store the customers of concern are the type of customers that visit the store regularly and purchase goods in a good quantity, therefore along with the avg price of product that a customer of a particular category buys , we also need to see that on an avg, how many transactions a customer of a particular category does , and also on an avg how many packets of chips does a customer of a particular category, buys in one transaction.

### First we have to find out that how many transactions are done by different category of customers 

```python
df_1=pd.DataFrame(df.groupby(['PREMIUM_CUSTOMER','LIFESTAGE'])['DATE'].count())
df_1.reset_index(inplace=True)
df_1.rename(columns={'DATE':'COUNT'},inplace=True)
df_1
```

### We also have to find out that how many different customers of a particular category have purchased chips.

```python
df_customer=pd.read_csv('QVI_purchase_behaviour.csv')
df_customer.head()
```

```python
df_2=pd.DataFrame(df_customer.groupby(['PREMIUM_CUSTOMER','LIFESTAGE'])['LYLTY_CARD_NBR'].count())
df_2.reset_index(inplace=True)
df_2.rename(columns={'LYLTY_CARD_NBR':'COUNTS'},inplace=True)
df_2
```

### Finding the average number of transactions a customer of a particular category does.

```python
df_3=pd.DataFrame(df_1['COUNT']/df_2['COUNTS'])
df_3.rename(columns={0:'#Transactions_Per_customer'},inplace=True)
df_4=pd.concat([df_2,df_3],axis=1)
df_4.drop('COUNTS',axis=1,inplace=True)
df_4
```

### We also need to find out that on an average how many packets of chips does a customer of a particular category purchases in one transaction.

```python
avg_packets=pd.DataFrame(df.groupby(['PREMIUM_CUSTOMER','LIFESTAGE']).sum())
avg_packets.reset_index(inplace=True)
df_4['#packets_per_transaction']=avg_packets['PROD_QTY']/df_1['COUNT']
df_4
```

### Finding avg price of product that a customer of a particular category purchases.

```python
avg_price=pd.DataFrame(df.groupby(['PREMIUM_CUSTOMER','LIFESTAGE']).sum())
avg_price.reset_index(inplace=True)
df_4['avg_price_product']=avg_price['PricePer100g']/df_1['COUNT']
df_4
```

### Finding what contribution does each category of customers make in Total Sales of chips

```python
df_5=pd.DataFrame(df.groupby(['PREMIUM_CUSTOMER','LIFESTAGE'])['TOT_SALES'].sum())
df_5=df_5.reset_index()
df_5['PERCENT_OF_TOT']=(df_5.TOT_SALES/df.TOT_SALES.sum())*100
df_5
```

```python
df_4['Percent_of_TotSales']=df_5['PERCENT_OF_TOT']
df_4.head()
```

```python
df_4.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
temp=scaler.fit_transform(df_4[["#Transactions_Per_customer","#packets_per_transaction","avg_price_product"]])
temp
```

```python
df_4["#Transactions_Per_customer_"]=temp[:,0]
df_4["#packets_per_transaction_"]=temp[:,1]
df_4["avg_price_product_"]=temp[:,2]
df_4
```

### Regular customers that purchases chips of reasonable price in good quantity are of greater importance than customers that visit the super-market less frequently and purchases fewer packets of expensive chips.

```python
df_4['Score']=df_4['#Transactions_Per_customer_']+df_4['#packets_per_transaction_']+df_4['avg_price_product_']
df_4
```

```python
df_4.describe()
```

```python
df_4.sort_values(by="Score", ascending=False).loc[df_4["Score"]>1]
```

### Percent_of_TotSales is supposed to be highly coorelated with   #Transactions_Per_customer,  #packets_per_transaction, avg_price_product.


```python
df_main_customers=df_4.sort_values(by="Score", ascending=False).loc[df_4["Score"]>1]
df_main_customers.reset_index(inplace=True)
df_main_customers.drop('index',axis=1,inplace=True)
df_main_customers
```

```python
plt.figure(figsize=(35,30))

for i in range(df_main_customers.shape[0]):
    lst=[]
    for j in df.PROD_FLAVOUR.unique():
        num=df[(df.PREMIUM_CUSTOMER==df_main_customers.iloc[i].PREMIUM_CUSTOMER)&(df.LIFESTAGE==df_main_customers.iloc[i].LIFESTAGE)&(df.PROD_FLAVOUR==j)]['PROD_QTY'].sum()
        lst.append(num)
    plt.subplot(5,2,i+1)
    plt.pie(lst,labels=list(df.PROD_FLAVOUR.unique()),autopct='%.2f%%')
    plt.title(df_main_customers.iloc[i].PREMIUM_CUSTOMER+' '+df_main_customers.iloc[i].LIFESTAGE)
plt.show()
```

```python
df_main_customers["Percent_of_TotSales"].sum()
```

### We can see that Young and Older families are the majors chips buyers , and by above pie-charts we can also say that all imp categories of customers are fond of flavours like Cheese, Chilli, Salty, while Tomato or tangy flavour is less purchased .

### The above 9 categories of customers have more than 53% of share in Total sales of chips.

### Therefore the super-market needs to maintain their chips stock according to the choice of their major chip buying customer categories.
