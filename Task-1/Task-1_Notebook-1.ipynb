### Trying to extract flavour of chips from PROD_NAME 

```python
lst=list(df1.PROD_NAME)
```

```python
ls=[]

for i in lst:
    lst1=re.findall(r'[cC][hH]+[iI]*[cC]*[kK]{0,2}[ieIE]*[nN]',i)
    if len(lst1)==0:
        lst1=re.findall(r'[Ss][oO][uU][rR]\s?|[Cc][rR][eE]?[aA]?[mM]\s?\&|\&?[Oo][nN][iI][oO]?[nN]',i)
    else:
        ls.append('Chicken_flavour')
        continue
    if len(lst1)==0:
        lst1=re.findall(r'[cC][hH]+[iI]*[lL]+[ieIEyY]+|[Ss][pP][iI]*[Cc][yYiIeE]+|[Jj][eaA]*[lL]+[eaA]*[pP]+[iIeEn]+[n]*[o]*',i)
    else:
        ls.append('Sour_Cream/Onion')
        continue
    if len(lst1)==0:
        lst1=re.findall(r'[cC][hH]+[eE]*[zZsS]+[eyEY]+|[Mm][oO]?[z]+[a]?[r]+[e]*[l]+[a]+',i)
    else:
        ls.append('Chilli')
        continue
    if len(lst1)==0:
        lst1=re.findall(r'[Ss][Aa]*[Ll][tT][yY]*',i)
    else:
        ls.append('Cheese')
        continue
    if len(lst1)==0:
        lst1=re.findall(r'[Tt][oO]?[mM][aA]?[tT][oO]|[Tt][aA][nN][gG][iIyY]',i)
    else:
        ls.append('Salty')
        continue
    if len(lst1)==0:
        ls.append('Other_flavour')
    else:
        ls.append('Tomato/Tangy')
```

```python
PROD_FLAVOUR=pd.Series(ls)
```

```python
PROD_FLAVOUR.value_counts()
```

```python
df1['PROD_FLAVOUR']=PROD_FLAVOUR
df1.head(10)
```
